# Question 12 - Leaders in hits from 1940 up to and including 1949. (Dataframe)

```python
years = np.linspace(1940, 1949,10).astype(np.int32)
df = pd.read_csv('Batting.csv')
df_ldrs = {}
cnt = 0
for i in years:
    temp = df[df['yearID'] == i].groupby('playerID').sum().sort_values(by = 'H', ascending=False).head(1)
    df_ldrs[cnt] = pd.DataFrame(temp)
    cnt = cnt + 1
result1 = pd.concat(df_ldrs)
result1
```

# Question 13 - Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number)

```python
df = pd.read_csv('Batting.csv')
years = df['yearID'].unique()
df_ldrs = {}
cnt = 0
for i in years:
    temp = df[df['yearID'] == i].groupby('playerID').sum().sort_values(by = 'H', ascending=False).head(1)
    df_ldrs[cnt] = pd.DataFrame(temp)
    cnt = cnt + 1
result1 = pd.concat(df_ldrs)
res = result1
res.reset_index(inplace=True)
res1 = res.groupby('playerID')['H'].sum()
res1 = pd.DataFrame(res1)
res2 = res.groupby('playerID')['HR'].count()
res2 = pd.DataFrame(res2)
res3 = pd.DataFrame()
res3['Hits'] = res1['H']
res3['Count'] = res2['HR']
res3 = res3.sort_values(by= 'Count', ascending=False)
res3.head(1)
```

# Question 25 - Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe)

```python
batE['age']=batE['yearID']-batE['birthYear']
player_age = {}
a = batE.groupby(by=['playerID'],as_index=False)[['HR']].sum()
a = a[a['HR']>=300]
for player in a["playerID"].tolist():
    players = batE[batE["playerID"] == player]
    cnt = 0
    for entry in players.iterrows():
        cnt += entry[1]["HR"]
        if cnt>=300:
            player_age[player] = entry[1]["age"]
            break

sorted_age = sorted(player_age.items(),key=lambda x:x[1])
pd.DataFrame(sorted_age[:5],columns=["playerID","age"])
```
